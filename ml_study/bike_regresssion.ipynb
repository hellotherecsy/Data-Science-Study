```python
#feature_names = ["season","holiday","workingday","weather" , "temp", "atemp","humidity","windspeed"]
#feature_names = ["season","workingday", "humidity"]
#feature_names = ["season","workingday", "humidity","month","hour"] # 82 
#feature_names = ["season","workingday", "humidity","month","hour","atemp"] # 81 
#feature_names = ["season","workingday", "humidity","month","hour","atemp","year"] # 43
#feature_names = ["season","workingday", "humidity","month","hour","year","windspeed"] # 49
#feature_names = ["season","workingday","humidity","month","hour","atemp","year","dayofweek"] # 41
feature_names = ["season","workingday","humidity","month","hour","atemp","year","dayofweek","windspeed"] # 41
```

```python
X_train = train[feature_names]
print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names]
print(X_test.shape)
X_test.head()
```

```python
label_name= "count"
y_train = train[label_name]
print(y_train.shape)
y_train.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=3)
model
```

# Score 

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(model, X_train, y_train, cv = 20,
                       scoring="neg_mean_absolute_error").mean()
print("Score = {0:.5f}".format(score))
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```
