## 컬럼 추가 수정 

```python
train["Category"] = "Titanic"
```

```python
train.head()
```

```python
train.shape
```

```python
train["Id"] = range(0,891)
```

```python
train.shape
```

```python
train.head()
```

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
train[["FamilySize", "SibSp" ,"Parch"]].head(10)
```

```python
train["Nationality_FR"] = train["Embarked"] == "C"
train["Nationality_UK"] = train["Embarked"].isin(["S","Q"])
```

```python
train[["Embarked","Nationality_FR", "Nationality_UK"]].head()
```

```python
train.loc[train["Embarked"] == "C" , "Nationaltity"] = "France"
train.loc[train["Embarked"].isin(["S","Q"]) , "Nationaltity"] = "England"
train.head()
```

```python
train["LowFare"] =    train["Fare"] < 15
train["MediumFare"] = (train["Fare"] >= 15) & (train["Fare"] < 35)
train["HighFare"] =  (train["Fare"] > 35)
```

```python
train[["Fare","LowFare","MediumFare","HighFare"]].head()
```

```python
mean_age = train["Age"].mean()
```

```python
train.loc[train["Age"].isnull() , "Age" ] = mean_age
```
